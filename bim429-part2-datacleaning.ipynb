```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data=pd.read_csv("/kaggle/input/sondata/birlestirilimis.csv")
```

```python
data.head(10)
```

```python
unwanted_genres = ['Biography', 'Reality-TV', 'Game-Show', 'yok', 'Documentary', 'Talk-Show', 'News', 'Sport', 'Music', 'Musical','Short']
data = data.dropna(subset=['Genre'])
data['Genre'] = data['Genre'].apply(lambda x: [] if pd.isna(x) or (isinstance(x, list) and len(x) == 0) else x)
data = data[~data['Genre'].apply(lambda x: any(genre in x for genre in unwanted_genres))]
```

```python
data = data.dropna(subset=['Genre'])
```

```python
data = data.dropna(subset=['Actors'])  # 'Actors' sütununda nan değer içeren satırları sil
#yok yazanları silemedimm bi de [] böyle olanları  
data = data[~data['Actors'].apply(lambda x: x == 'yok' if isinstance(x, str) else False)]  # 'yok' yazan satırları sil
# Liste boşsa satırı sil
data = data[data['Actors'].apply(lambda x: bool(x))]
```

```python
data = data[data['Genre'] != 'Comedy']
```

```python
data = data[data['Name'] != 'yok']
```

```python
data['Season'] = pd.to_numeric(data['Season'], errors='coerce')
data['Season'] = data['Season'].fillna(1)
data['Season'] = data['Season'].apply(lambda x: 1 if x > 35 else x)
```

```python
data['Season'].unique()
```

```python
data['Episode'] = pd.to_numeric(data['Episode'], errors='coerce')
data['Episode'] = data['Episode'].fillna(data['Episode'].mean()*1/2).astype(int)
```

```python
data.tail(30)
```

```python
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')

# 'Rating' sütunundaki boş değerleri bütün sütunun ortalaması ile doldurun
data['Rating'] = data['Rating'].fillna(data['Rating'].mean())

# Virgülden sonra sadece bir hane yazma
data['Rating'] = data['Rating'].apply(lambda x: round(x, 1));
```

```python
data.tail()
```

```python
data = data.dropna(subset=['Directors'])
data = data[~data['Directors'].apply(lambda x: x == 'yok' if isinstance(x, str) else False)]
```

```python
data = data.dropna(subset=['Writers'])
data = data[~data['Writers'].apply(lambda x: x == 'yok' if isinstance(x, str) else False)]
```

```python
data['AvgTime'] = data['AvgTime'].apply(lambda x: '0h' if 'h' not in x and 'm' not in x else x)
```

```python
data['AvgTime'] = data['AvgTime'].fillna("0h");
```

```python
data = data[data['AvgTime'].str.contains('m|h', regex=True, na=False)]
```

```python
data['AvgTime'].unique()
```

```python
data['Hours'] = data['AvgTime'].str.extract('(\d+)h').fillna(0).astype(int)
data['Minutes'] = data['AvgTime'].str.extract('(\d+)m').fillna(0).astype(int)

# Toplam süre sütunu oluşturma (saatleri dakikaya çevirme ve dakikaları ekleme)
data['Runtime'] = data['Hours'] * 60 + data['Minutes']
```

```python
data = data.drop(['Hours', 'Minutes'], axis=1)
```

```python
data=data.drop('AvgTime', axis=1)
```

```python
data['Runtime'] = data['Runtime'].replace(0, np.nan)
data['Runtime'] = data['Runtime'].fillna(data['Runtime'].mean()*2/3).astype(int)
```

```python
data.tail(30)
```

```python
data['Date'] = data['Date'].astype(str)
data['IsFinished'] = data['Date'].apply(lambda x: 0 if len(x) == 5 else 1)
```

```python
data=data.dropna(subset=['Date'])
data = data[~data['Date'].apply(lambda x: x == 'nan' if isinstance(x, str) else False)]
```
